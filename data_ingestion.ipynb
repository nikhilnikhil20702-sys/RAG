```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader('myfood.txt',encoding="utf-8")
text_docs = loader.load()


```

```python
text_docs[0].page_content
```

```python
# how to reccursively split text by characters
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=50,chunk_overlap=20)
final_docs =text_splitter.split_documents(text_docs)
```

```python
print(final_docs[0].page_content)
print(final_docs[1].page_content)
```

```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings=(OllamaEmbeddings(model="embeddinggemma:latest"))
```

```python
from langchain_community.vectorstores import FAISS
```

```python
embeddings.embed_query("hi")
```

```python
from langchain_community.vectorstores import FAISS
```

```python
db = FAISS.from_documents(final_docs,embedding=embeddings)
```

```python
db.save_local("faiss_index")  ##saving and loading
```

```python
query ="what are the cuisine available here?" 
db.similarity_search(query=query)
```
